```python
# -- 将数据框命名为crime
import pandas as pd 
import numpy as np 
crime = pd.read_csv('US_Crime_Rates_1960_2014.csv',index_col=0)
crime.sample(5)
```

```python
# -- 每一列(column)的数据类型是什么样的？
crime.info()
```

```python
# -- 将Year的数据类型转换为 datetime64
crime['Year'] = pd.to_datetime((crime['Year'].apply(lambda x:str(x))))
crime.sample(5)
```

```python
# -- 将列Year设置为数据框的索引
crime.set_index('Year',drop=True,inplace=True)
```

```python
crime.sample(5)
```

```python
# -- 删除名为Total的列
crime.drop(['Total'],axis=1,inplace=True)
crime.sample(5)
```

```python
# -- 按照Year（每十年）对数据框进行分组并求和
crime_10y = crime.resample('10y').sum()
crime_10y 
```

```python
# -- 何时是美国历史上生存最危险的年代？
crime_10y.max()
```
