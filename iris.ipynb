```python
# -- 将数据集存成变量iris
import pandas as pd 
iris = pd.read_csv('iris.data',header=None)
iris 
```

```python
# -- 创建数据框的列名称['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris.columns = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris 
```

```python
# -- 数据框中有缺失值吗？
iris.isnull().sum()
```

```python
iris.info()
```

```python
# -- 将列petal_length的第10到19行设置为缺失值
from numpy import nan as NaN
iris.petal_length[10:20] = NaN
iris.petal_length.head(21)
```

```python
# -- 将petal_lengt缺失值全部替换为1.0
iris.fillna(1.0,inplace=True)
iris.petal_length.head(21)
```

```python
# -- 删除列class
iris.drop(['class'],axis=1,inplace=True)
iris 
```

```python
# -- 将数据框前三行设置为缺失值
iris.iloc[0:3,:] =  NaN
iris.iloc[0:5,:]
```

```python
# -- 删除有缺失值的行
iris.dropna(inplace=True)
iris 
```

```python
# -- 重新设置索引
iris.index = range(0,len(iris))
iris 
```
