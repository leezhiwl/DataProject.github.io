```python
# -- 将数据作存储并且设置前三列为合适的索引
import pandas as pd 
wind = pd.read_csv('wind.csv',sep='\s+',parse_dates=[[0,1,2]])
wind.sample(5)
```

```python
import datetime
def fix_year(x):
    year = x.year - 100 if x.year>1989 else x.year
    return datetime.date(year,x.month,x.day)
wind['Yr_Mo_Dy'] =  wind['Yr_Mo_Dy'].apply(fix_year)   
wind
```

```python
wind.info()
```

```python
# -- 将日期设为索引，注意数据类型，应该是datetime64[ns]
wind['Yr_Mo_Dy'] = pd.to_datetime(wind['Yr_Mo_Dy'])
wind = wind.set_index('Yr_Mo_Dy')
wind 
```

```python
wind.index
```

```python
# -- 对应每一个location，一共有多少数据值缺失
wind.isnull().sum()
```

```python
# -- 对应每一个location，一共有多少完整的数据值
wind.shape[0] - wind.isnull().sum()
```

```python
# -- 对于全体数据，计算风速的平均值
wind.mean().mean()
```

```python
wind.sum().sum()/(wind.shape[0]*wind.shape[1])
```

```python
# -- 创建一个名为loc_stats的数据框去计算并存储每个location的风速最小值，最大值，平均值和标准差
loc_stats = wind.agg(['min','max','mean','std'])
loc_stats 
```

```python
# -- 创建一个名为day_stats的数据框去计算并存储所有location的风速最小值，最大值，平均值和标准差
day_stats = wind.agg(['min','max','mean','std'],axis=1)
day_stats
```

```python
# -- 对于每一个location，计算一月份的平均风速
wind[wind.index.month==1].mean()
```

```python
# -- 对于数据记录按照年为频率取样
wind.resample('1Y').mean()
```
